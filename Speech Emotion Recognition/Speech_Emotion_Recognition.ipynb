```python
emotions = {'01' : 'neutral', '02' : 'calm', '03' : 'happy', '04' : 'sad', '05' : 'angry', '06' : 'fearful', '07' : 'disgust', '08' : 'surprised'}
```

```python
def extract_features(path):
    try:
        sig, rate = alib.load(path)
            
        mfcc = alib.feature.mfcc(sig, rate)
        mfcc_mean = mfcc.mean(axis = 1).tolist()
        mfcc_var = mfcc.var(axis = 1).tolist()
            
        ch_stft = alib.feature.chroma_stft(sig, rate)
        chroma_stft_mean = ch_stft.mean(axis = 1).tolist()
        chroma_stft_var = ch_stft.var(axis = 1).tolist()
        
        spectral_centroid = alib.feature.spectral_centroid(sig, rate)
        spectral_centroid_mean = spectral_centroid.mean()
        spectral_centroid_var = spectral_centroid.var()
            
        spectral_bandwidth = alib.feature.spectral_bandwidth(sig, rate)
        spectral_bandwidth_mean = spectral_bandwidth.mean()
        spectral_bandwidth_var = spectral_bandwidth.var()
            
        rms = alib.feature.rms(sig)
        rms_mean = rms.mean()
        rms_var = rms.var()
            
        zero_crossing_rate = alib.feature.zero_crossing_rate(sig)
        zero_crossing_rate_mean = zero_crossing_rate.mean()
        zero_crossing_rate_var = zero_crossing_rate.var()
        
        return mfcc_mean + mfcc_var + chroma_stft_mean + chroma_stft_var + [spectral_centroid_mean, spectral_centroid_var,
                                                                            spectral_bandwidth_mean, spectral_bandwidth_var,
                                                                            rms_mean, rms_var, zero_crossing_rate_mean, 
                                                                            zero_crossing_rate_var]
    except:
        pass
```

```python
x = []
y = []
for folder in os.listdir(directory):
    print(folder)
    for file in os.listdir(directory + '/' + folder):
        print(folder, file)
        try:
            emo = emotions.get(file.split('-')[2])
            path = ('/'.join([directory, folder, file]))
            features = extract_features(path)
            x.append(features)
            y.append(emo)
        except:
            pass
print('Done')

x = pd.DataFrame(x)
y = pd.Series(y)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.25, random_state = 101)
```

```python
model_rf = RandomForestClassifier().fit(xtrain, ytrain)
print(round(100*model_rf.score(xtest, ytest),2))
print(round(100*model_rf.score(xtrain, ytrain),2))
```
