### Filling missing data

```python
def age_treatment(x):
    if(pd.isnull(x['Age'])):
        return float(df['Age'][df['ProductPitched']==x['ProductPitched']].mean())
    return x['Age']


def num_of_followups_treatment(x):
    if(pd.isnull(x['NumberOfFollowups'])):
        return float(df['NumberOfFollowups'][(df['NumberOfPersonVisiting']==x['NumberOfPersonVisiting'])].mode())
    return x['NumberOfFollowups']
```

```python
df['Age'] = df[['Age', 'ProductPitched']].apply(age_treatment, axis = 1)

df['TypeofContact'].replace(to_replace=np.nan, value = df['TypeofContact'].mode(), inplace=True)

df['DurationOfPitch'].replace(to_replace=np.nan, value = df['DurationOfPitch'].median(), inplace=True)

df['NumberOfFollowups'] = df[['NumberOfFollowups', 'NumberOfPersonVisiting']].apply(num_of_followups_treatment, axis = 1)

df['NumberOfTrips'].replace(to_replace=np.nan, value = df['NumberOfTrips'].median(), inplace=True)

df = df.dropna(axis = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

def feature_scaling(x):
    scaled_features = sc.fit_transform(x)
    x_df = pd.DataFrame(scaled_features, index = x.index, columns = x.columns)
    return x_df
```

## Handling Imbalance Data using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state = 0)

def smote(x, y):
    print("Before OverSampling, counts of label '1': {}".format(sum(y==1)))
    print("Before OverSampling, counts of label '0': {} \n".format(sum(y==0)))

    x, y = sm.fit_resample(x, y.ravel())

    print("After OverSampling, counts of label '1': {}".format(sum(y==1)))
    print("After OverSampling, counts of label '0': {}".format(sum(y==0)))
    return x, y
```

```python
x, y = smote(x, y)
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
def split_data(x, y):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = .25, random_state = 0)
    return xtrain, xtest, ytrain, ytest
```

```python
xtrain, xtest, ytrain, ytest = split_data(x, y)
```

## Prediction

### Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

#### Selecting no. of neighbors for KNN

```python
fig = plt.subplots(figsize = (18,6))

error1= []
error2= []

for k in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain, ytrain)
    
    ypred1 = knn.predict(xtrain)
    error1.append(np.mean(ytrain != ypred1))
    
    ypred2 = knn.predict(xtest)
    error2.append(np.mean(ytest!= ypred2))

plt.plot(range(1,20),error1,label="train")
plt.plot(range(1,20),error2,label="test")
plt.xlabel('k Value')
plt.ylabel('Error')
plt.legend()
```

We will not be using KNN

```python
models = []

models.append(['Logistic Regression', LogisticRegression()])
models.append(['Decision Tree', DecisionTreeClassifier()])
models.append(['Random Forest', RandomForestClassifier()])
models.append(['XG Boost', GradientBoostingClassifier()])
```

#### Analysis of score and evaluation metrics

```python
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn.model_selection import cross_val_score, KFold
```

```python
def analyze(models, xtrain, ytrain, xtest, ytest, x, y):
    for m in range(len(models)):
        model = models[m][1]

        #training the model
        model.fit(xtrain, ytrain)

        #test score and training score
        testing_score = model.score(xtest, ytest)
        training_score = model.score(xtrain, ytrain)

        #K-Fold cross validation mean accuracy
        kfold = KFold(n_splits = 10, random_state = 0, shuffle=True)
        kfold_accuracy = cross_val_score(model, x, y, cv = kfold, scoring = 'accuracy')
        kfold_mean_accuracy = kfold_accuracy.mean()
        std = kfold_accuracy.std()

        ypred = model.predict(xtest)
        
#       precision
        precision = precision_score(ytest, ypred)

        #Confusion Matrix
        cm = confusion_matrix(ytest, ypred)

        #Classification Report
        cr = classification_report(ytest, ypred)

        #Features Importances
        if(models[m][0] == 'Logistic Regression'):
            pass
        else:
            feature_imp = pd.Series(model.feature_importances_*100, index=x.columns).sort_values(ascending=False)

        #Printing the values and plotting the graph
        print(models[m][0] + ': ' )
        print()
        print('Training score => ' + str(round(training_score*100, 2)) + '%')
        print('Testing score => ' + str(round(testing_score*100, 2)) + '%')
        print('K-Fold mean accuracy => ' + str(round(kfold_mean_accuracy*100, 2)) + '%')
        print('Standard Deviation in kfold => ' + str(round(std, 4)))
        print()
        print('Precision score =>' + str(round(precision*100, 2)) + '%')
        print()
        print(cm)
        print()
        print(cr)
        print()
        if(models[m][0] == 'Logistic Regression'):
            pass
        else:
            sns.barplot(x = feature_imp, y = feature_imp.index)
            plt.xlabel("Features Importance")
            plt.ylabel("Features")
            plt.show()
            print()
        print('*'*90)
```

```python
analyze(models, xtrain, ytrain, xtest, ytest, x, y)
```
